```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 120, 500)  
z_position = np.linspace(1, 1, 500)  
x_position = np.linspace(0, 0, 500) 
y_position = np.linspace(0, 0, 500) 

plt.figure(figsize=(8, 5))
plt.plot(time, z_position, color='blue', label="Z Position")
plt.plot(time, x_position, color='black', label="X Position")
plt.plot(time, y_position, color='black', label="Y Position")

plt.xlim(0, 125) 
plt.ylim(-0.5, 1.5)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Drone Position vs. Time")

plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 120, 500) 
z_position = np.linspace(0, 0, 500) 
x_position = np.linspace(0, 0, 500) 
y_position = np.linspace(0, 0, 500) 

plt.figure(figsize=(8, 5))
plt.plot(time, z_position, color='black', label="φ")
plt.plot(time, x_position, color='black', label="\u03B8")
plt.plot(time, y_position, color='black', label="\u03C8")

plt.xlim(0, 125) 
plt.ylim(-90, 180)
plt.xlabel("Time (s)")
plt.ylabel("Attitude (Degrees)")
plt.title("Drone Attitude vs. Time")

plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 120, 500)
z_position = np.linspace(1, 1, 500)
x_position = 2*np.cos(0.25*time)
y_position = 2*np.sin(0.25*time)

plt.figure(figsize=(8, 5))
plt.plot(time, z_position, color='blue', label="Z Position")
plt.plot(time, x_position, color='black', label="X Position")
plt.plot(time, y_position, color='red', label="Y Position")

plt.xlim(0, 125) 
plt.ylim(-10, 10)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Drone Position vs. Time")

plt.grid(True)
plt.legend()

plt.show()
```
